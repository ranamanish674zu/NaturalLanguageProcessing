#removing the stopword

```python

nltk.download('stopwords')
from nltk.corpus import stopwords
sw=set (stopwords.words('english'))
print(sw)#comman stopword in english 
```

```python
def remove_stopwords(text,stopwords):
  useful_words=[w for w in text if w not in stopwords]
  return useful_words
```

```python
text ="i am not bothered about her very much ".split()
useful_text=remove_stopwords(text ,sw)
print(useful_text)#remove all the stopword and printing all useful word
```

#TF -IDF NORMALIZATION 

##->Avoid features that occurs very often,becauase they contains less information
##->Information decreases as the numbers of occurences increases across different type of document 
##->so we define another term-term document-frequency which associates a weight with every word

```python
from nltk.corpus.reader.sentiwordnet import SentiSynset
#full form ->term frequency(how many time a word appears across the sentence) -inverse document frequency(how many time the word appears across the corpus )

sent_1="this is good movie"
sent_2="this was good movie"
sent_3="this is not good movie"
corpus=[sent_1,sent_2,sent_3]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
```

```python
vc=tfidf.fit_transform(corpus).toarray()
print(vc)
```
